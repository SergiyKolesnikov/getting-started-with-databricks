```
%sql select * from wine.normalized;
```

```
df = spark.sql("select normalized, quality from wine.normalized")
```

```
from pyspark.ml.regression import LinearRegression
lr = LinearRegression(labelCol="quality", featuresCol="normalized")

train, test = df.randomSplit([0.75, 0.25])

model = lr.fit(train)
```

```
ts = model.summary

print(f"RMSE: {ts.rootMeanSquaredError}")
print(f"R2: {ts.r2}")
```

```
from pyspark.ml.evaluation import RegressionEvaluator

eval = RegressionEvaluator(labelCol="quality", predictionCol="prediction")

pred = model.transform(test)

print(f"RMSE: {eval.evaluate(pred, {eval.metricName: 'rmse'})}")
print(f"R2: {eval.evaluate(pred, {eval.metricName: 'r2'})}")
```
